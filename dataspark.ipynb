```python
import pandas as pd 

customer_df=pd.read_csv('D:/DATA SCIENCE/Vs/.venv/Uncleaned_Data_set/Customers.csv',encoding='latin')
exchange_rate_df=pd.read_csv('D:/DATA SCIENCE/Vs/.venv/Uncleaned_Data_set/Exchange_Rates.csv',encoding='latin')
products_df=pd.read_csv('D:/DATA SCIENCE/Vs/.venv/Uncleaned_Data_set/Products.csv',encoding='latin')
sales_df=pd.read_csv('D:/DATA SCIENCE/Vs/.venv/Uncleaned_Data_set/Sales.csv',encoding='latin')
stores_df=pd.read_csv('D:/DATA SCIENCE/Vs/.venv/Uncleaned_Data_set/Stores.csv',encoding='latin')
```

```python
# need to check for the missing values
print("Missing values in Customers Dataset:\n", customer_df.isnull().sum(), "\n")
print("Missing values in Exchange Rates Dataset:\n", exchange_rate_df.isnull().sum(), "\n")
print("Missing values in Products Dataset:\n", products_df.isnull().sum(), "\n")
print("Missing values in Stores Dataset:\n", stores_df.isnull().sum(), "\n")
print("Missing values in Sales Dataset:\n", sales_df.isnull().sum(), "\n")
```

```python
# to handle the missing values
customer_df.ffill(inplace=True)
exchange_rate_df.ffill(inplace=True)
products_df.ffill(inplace=True)
sales_df.ffill(inplace=True)
stores_df.ffill(inplace=True)
```

```python
# to check for duplicates in the datasets
print("Duplicate rows in Customers Dataset:", customer_df.duplicated().sum())
print("Duplicate rows in Exchange Rates Dataset:", exchange_rate_df.duplicated().sum())
print("Duplicate rows in Products Dataset:", products_df.duplicated().sum())
print("Duplicate rows in Stores Dataset:", stores_df.duplicated().sum())
print("Duplicate rows in Sales Dataset:", sales_df.duplicated().sum())
```

```python
# Removing duplicates in the datasets
customer_df.drop_duplicates(inplace=True)
exchange_rate_df.drop_duplicates(inplace=True)
products_df.drop_duplicates(inplace=True)
stores_df.drop_duplicates(inplace=True)
sales_df.drop_duplicates(inplace=True)
```

```python
# Filter out rows where 'Zip Code' contains non-numeric values
customer_df = customer_df[customer_df['Zip_Code'].str.isnumeric()]
# Convert 'Zip Code' to integer
customer_df['Zip_Code'] = customer_df['Zip_Code'].astype('int32')
```

```python
customer_df['Birthday'] = pd.to_datetime(customer_df['Birthday'], errors='coerce', format='%m-%d-%Y')
customer_df['Birthday'] = customer_df['Birthday'].fillna(pd.to_datetime(customer_df['Birthday'], format='%m/%d/%Y', errors='coerce'))
```

```python
# Extract numeric part
products_df['Unit_Cost_USD'] = products_df['Unit_Cost_USD'].str.extract('(\d+)')
# Convert to integer, handle NaN by dropping them or filling them
products_df['Unit_Cost_USD'] = products_df['Unit_Cost_USD'].dropna().astype('int32')
```

```python
# Extract numeric part
products_df['Unit_Price_USD'] = products_df['Unit_Price_USD'].str.extract('(\d+)')
# Convert to integer, h-andle NaN by dropping them or filling them
products_df['Unit_Price_USD'] = products_df['Unit_Price_USD'].dropna().astype('int32')
```

```python
# Parse the Order_Date column with multiple formats
sales_df['Order_Date'] = pd.to_datetime(sales_df['Order_Date'], errors='coerce', format='%m-%d-%Y')
sales_df['Order_Date'] = sales_df['Order_Date'].fillna(pd.to_datetime(sales_df['Order_Date'], format='%m/%d/%Y', errors='coerce'))

# Parse the Delivery_Date column with multiple formats
sales_df['Delivery_Date'] = pd.to_datetime(sales_df['Delivery_Date'], errors='coerce', format='%m-%d-%Y')
sales_df['Delivery_Date'] = sales_df['Delivery_Date'].fillna(pd.to_datetime(sales_df['Delivery_Date'], format='%m/%d/%Y', errors='coerce'))
```

```python
exchange_rate_df['Date'] = pd.to_datetime(exchange_rate_df['Date'])
```

```python
# Extracting year and month from the date
exchange_rate_df['Year'] = exchange_rate_df['Date'].dt.year 
exchange_rate_df['Month'] = exchange_rate_df['Date'].dt.month
```

```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# MySQL credentials
mysql_user = 'root'
mysql_password = 'root'
mysql_host = '127.0.0.1'
mysql_port = '3306'
mysql_database = 'Dataspark'

# Create a connection to MySQL database
engine = create_engine(f'mysql+pymysql://{mysql_user}:{mysql_password}@{mysql_host}:{mysql_port}/{mysql_database}', echo=True)
```

```python
#load the cleaned datasets into sql
customer_df.to_sql('customer', con=engine, index=False, if_exists='replace')
exchange_rate_df.to_sql('exchange_rate', con=engine, index=False, if_exists='replace')
products_df.to_sql('products', con=engine, index=False, if_exists='replace')
stores_df.to_sql('stores', con=engine, index=False, if_exists='replace')
sales_df.to_sql('sales', con=engine, index=False, if_exists='replace')

engine.dispose()
```

```python
#saving the cleaned & pushed dataset for backup
customer_df.to_csv('customers.csv', index=False)
exchange_rate_df.to_csv('exchange_rates.csv', index=False)
products_df.to_csv('products.csv', index=False)
stores_df.to_csv('stores.csv', index=False)
sales_df.to_csv('sales.csv', index=False)
```
